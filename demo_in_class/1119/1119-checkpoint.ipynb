```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
X = np.random.randn(200,2)
X = np.matmul(X,[[5,0],[0,1]])

plt.plot(X[:,0],X[:,1],"b.")
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=1,random_state=42)
X_trans = pca.fit_transform(X)
```

```python
X_trans
```

```python
plt.plot(X_trans,np.zeros((200,1)),"b.")
```

```python
pca.components_[0]
```

```python
X_trans2 = np.matmul(X,pca.components_[0])
```

```python
plt.plot(X_trans2,np.zeros((200,1)),"b.")
```

```python
from sklearn.datasets import make_swiss_roll
from mpl_toolkits.mplot3d import Axes3D
```

```python
X, t= make_swiss_roll(n_samples=1000,noise=0.1,random_state=42)
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(X[:,0],X[:,1],X[:,2], c=t, cmap= plt.cm.hot)
ax.view_init(10,70)
```

```python
pca = PCA(n_components=2,random_state=42)
X_trans = pca.fit_transform(X)
plt.scatter(X_trans[:,0],X_trans[:,1], c=t, cmap= plt.cm.hot)
```

```python
from sklearn.manifold import LocallyLinearEmbedding

lle = LocallyLinearEmbedding(n_neighbors=10,n_components=2,random_state=42)
X_trans = lle.fit_transform(X)
plt.scatter(X_trans[:,0],X_trans[:,1], c=t, cmap= plt.cm.hot)
```

```python
from sklearn.manifold import MDS

mds = MDS(n_components=2,random_state=42)
X_trans = mds.fit_transform(X)
plt.scatter(X_trans[:,0],X_trans[:,1], c=t, cmap= plt.cm.hot)
```

```python
from sklearn.manifold import Isomap

iso = Isomap(n_neighbors=10,n_components=2)
X_trans = iso.fit_transform(X)
plt.scatter(X_trans[:,0],X_trans[:,1], c=t, cmap= plt.cm.hot)
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2,random_state=42)
X_trans = tsne.fit_transform(X)
plt.scatter(X_trans[:,0],X_trans[:,1], c=t, cmap= plt.cm.hot)
```
