```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
sites = gpd.read_file("data\SJER_plot_centroids.shp")
```

```python
sites.head(3)
sites.shape
```

```python
print(sites.crs) 
# Question 1
```

```python
crop = gpd.read_file("data\SJER_crop.shp")
crop.shape 
# Question 2 : there is 1 record in "SJER_crop.shp with 2 attributes."
```

```python
fig, axe = plt.subplots()
crop.plot(ax = axe, color = "black")
sites.plot(ax =axe, color = "white" ) 
# Question 3
```

```python
sites_buffer = sites.geometry.buffer(150) 
fig,axe2 = plt.subplots() 
sites.plot(ax = axe2, color = "black", markersize = 1 ) 
sites_buffer.plot(ax = axe2, cmap = plt.get_cmap("Set1"), alpha = 0.6)
# Question 4
```

```python
sites_buffer_gdf = gpd.GeoDataFrame(geometry = sites_buffer, crs = sites.crs)
areas_not_within = gpd.overlay(crop,sites_buffer_gdf,how='difference')
areas_not_within.plot()
# Question 5
```
