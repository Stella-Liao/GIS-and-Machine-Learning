```python
import pandas as pd
import numpy as np
```

```python
raw_cl = pd.read_csv('crime_landUse_MAN.csv')
raw_cl
```

```python
from sklearn import preprocessing
#label = preprocessing.LabelEncoder()
#crime = label.fit_transform(raw_cl.CrimeType)
days = pd.get_dummies(raw_cl.DayofWeek)
landuses = pd.get_dummies(raw_cl.LandUse)
times = pd.get_dummies(raw_cl.TimeRange)
cl=pd.concat([landuses,times,days,raw_cl.CrimeTypeID],axis=1)
#cl['crime']=crime
```

```python
cl
```

```python
from sklearn.model_selection import train_test_split
training,test=train_test_split(cl,train_size=0.8,random_state=42)
```

```python
train,validation=train_test_split(new_training,train_size=0.8,random_state=42)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import log_loss
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
feature_list = train.columns.tolist()
feature_list = feature_list[:len(feature_list) - 1] 
```

```python
#knn
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier()
model_knn.fit(train[feature_list],train.CrimeTypeID)
pred_knn_value = np.array(model_knn.predict(test[feature_list]))
predicted_knn = np.array(model_knn.predict_proba(test[feature_list]))
```

```python
log_loss(test.CrimeTypeID,predicted_knn),confusion_matrix(test.CrimeTypeID,pred_knn_value),f1_score(test.CrimeTypeID,pred_knn_value,average="weighted")
```

```python
from sklearn.naive_bayes import BernoulliNB
model_NB = BernoulliNB()
model_NB.fit(train[feature_list], train.CrimeTypeID)
pred_NB_value = np.array(model_NB.predict(test[feature_list]))
predicted_NB = np.array(model_NB.predict_proba(test[feature_list]))
log_loss(test.CrimeTypeID,predicted_NB),confusion_matrix(test.CrimeTypeID,pred_NB_value),f1_score(test.CrimeTypeID,pred_NB_value,average="weighted")
```

```python
from sklearn.linear_model import LogisticRegression
model_LR = LogisticRegression()
model_LR.fit(train[feature_list], train.CrimeTypeID)
pred_LR_value = np.array(model_LR.predict(test[feature_list]))
predicted_LR = np.array(model_LR.predict_proba(test[feature_list]))
log_loss(test.CrimeTypeID, predicted_LR),confusion_matrix(test.CrimeTypeID,pred_LR_value),f1_score(test.CrimeTypeID,pred_LR_value,average="weighted")
```

```python
from sklearn.ensemble import RandomForestClassifier  
model_RF = RandomForestClassifier()  
model_RF.fit(train[feature_list], train.CrimeTypeID)
pred_RF_value = np.array(model_RF.predict(test[feature_list]))
predicted_RF = np.array(model_RF.predict_proba(test[feature_list]))
log_loss(test.CrimeTypeID, predicted_RF),confusion_matrix(test.CrimeTypeID,pred_RF_value),f1_score(test.CrimeTypeID,pred_RF_value,average="weighted")
```

```python
import lightgbm as lgb
model_lgb =lgb.LGBMClassifier(objective='multiclass', num_class=3,random_state=42).fit(train[feature_list], train.CrimeTypeID)
pred_lgb_value = np.array(model_lgb.predict(test[feature_list]))
predicted_lgb = np.array(model_lgb.predict_proba(test[feature_list]))
log_loss(test.CrimeTypeID, predicted_lgb),confusion_matrix(test.CrimeTypeID,pred_lgb_value),f1_score(test.CrimeTypeID,pred_lgb_value,average="weighted")
```
