```python
import pandas as pd
import numpy as np
```

```python
raw_cl = pd.read_csv('crime_landUse_MAN.csv')
raw_cl
```

```python
from sklearn import preprocessing
#label = preprocessing.LabelEncoder()
#crime = label.fit_transform(raw_cl.CrimeType)
days = pd.get_dummies(raw_cl.DayofWeek)
landuses = pd.get_dummies(raw_cl.LandUseID)
times = pd.get_dummies(raw_cl.Time)
cl=pd.concat([landuses,times,raw_cl.CrimeTypeID],axis=1)
#cl['crime']=crime
```

```python
from sklearn.model_selection import train_test_split
training,test=train_test_split(cl,train_size=0.8,random_state=42)
```

```python
train,validation=train_test_split(new_training,train_size=0.8,random_state=42)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import log_loss
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
feature_list = train.columns.tolist()
feature_list = feature_list[:len(feature_list) - 1] 
```

```python
import lightgbm as lgb
model_lgb =lgb.LGBMClassifier(objective='multiclass', num_class=3,random_state=42).fit(train[feature_list], train.CrimeTypeID)
pred_lgb_value = np.array(model_lgb.predict(test[feature_list]))
predicted_lgb = np.array(model_lgb.predict_proba(test[feature_list]))
log_loss(test.CrimeTypeID, predicted_lgb),confusion_matrix(test.CrimeTypeID,pred_lgb_value),f1_score(test.CrimeTypeID,pred_lgb_value,average="weighted")
```

max_depth & num_leaves

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import log_loss
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {
    'max_depth': [4,6,8],
    'num_leaves': [20,30,40]
}

gbm = lgb.LGBMClassifier(learning_rate=0.1,num_iterations = 200,random_state=42)
gsearch1 = GridSearchCV(gbm, param_grid=parameters, scoring='neg_log_loss', cv=5)
gsearch1.fit(validation[feature_list], validation.CrimeTypeID)

print('参数的最佳取值:{0}'.format(gsearch1.best_params_))
print('最佳模型得分:{0}'.format(gsearch1.best_score_))
```

```python
import lightgbm as lgb
model_lgb =lgb.LGBMClassifier(objective='multiclass', 
                              num_class=3,
                              random_state=42,
                             max_depth=4,
                             num_leaves=20).fit(train[feature_list], train.CrimeTypeID)
pred_lgb_value = np.array(model_lgb.predict(test[feature_list]))
predicted_lgb = np.array(model_lgb.predict_proba(test[feature_list]))
log_loss(test.CrimeTypeID, predicted_lgb),confusion_matrix(test.CrimeTypeID,pred_lgb_value),f1_score(test.CrimeTypeID,pred_lgb_value,average="weighted")
```
